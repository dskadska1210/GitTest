```python
# Linear Regression
# 가장 기본적인 회귀모델로 입력 변수와 정답 사이의 관계를 직선으로 표현함
# 스케일링 : 권장
# 언제 사용할까
1. 구현하기 쉽기에 비교용으로 사용하기 좋음
2. 데이터의 관계가 단순할 경우
3. 각 변수가 결과에 얼마나 영향을 주는지 보기 좋음
from sklearn.linear_model import LinearRegression

modelLinear = LinearRegression(
    fit_intercept=True,   # True면 절편 b를 학습함. 일반적으로 True 사용
    n_jobs=None           # 사용할 CPU 개수. -1는 모든 CPU 사용
)

modelLinear.fit(xTrainScale, yTrain)

pred = modelLinear.predict(xTestScale)
```

```python
# Ridge Regression
# Linear Regression에 L2규제를 추가한 모델
# 가중치가 너무 커지지 않도록 막음
# 스케일링 : 권장
# 언제 사용할까
# 1. 선형회귀에서 과적합이 일어날 경우, 과적합 감소
# 2. 컬럼의 개수가 너무 많을 경우 가중치 제어
# 3. 다중공선성 완화
from sklearn.linear_model import Ridge

modelRidge = Ridge(
    alpha=1.0,            # 규제 강도. 클수록 가중치를 더 강하게 줄임
                          # 0.001, 0.01, 0.1, 1, 10, 100 ,1000 ...
    fit_intercept=True,   # 절편 학습 여부
    random_state=42       # 재현성 고정
)

modelRidge.fit(xTrainScale, yTrain)

pred = modelRidge.predict(xTestScale)
```

```python
# Lasso Regression
# Linear Regression에 L1규제를 추가한 모델
# 일부 가중치를 0으로 만들어, 중요하지 않은 변수 제거 가능
# 스케일링 : 권장
# 언제 사용할까
# 1. 변수가 너무 많을 경우, 중요하지 않는 변수를 0으로 만들어 제거 효과
# 2. 자동으로 중요한 변수를 선택하는 Feature Selection 효과
# 3. 해석이 필요한 모델이 필요한 경우 남은 변수로 해석 가능

from sklearn.linear_model import Lasso

modelLasso = Lasso(
    alpha=0.1,            # 규제 강도. 클수록 더 많은 계수가 0에 가까워짐
                          # 0.001, 0.01, 0.1, 1, 10, 100 ,1000 ...
    fit_intercept=True,   # 절편 학습 여부
    max_iter=10000,       # 최적화를 반복할 최대 횟수. 수렴 경고가 나면 증가시키기
    random_state=42       # 재현성 고정
)

modelLasso.fit(xTrainScale, yTrain)

pred = modelLasso.predict(xTestScale)
```

```python
# ElesticNet Regression
# Ridge와 Lasso를 섞은 모델
# 일부 가중치를 0으로 만들어, 중요하지 않은 변수 제거 가능
# 스케일링 : 권장
# 언제 사용할까
# 1. Ridge, Lasso중 고르기 애매할 경우. 둘의 장점이 섞여있음
# 2. 변수도 많고 변수간 상관관계가 있을 경우(다중공선성)

from sklearn.linear_model import ElasticNet

modelElastic = ElasticNet(
    alpha=0.1,            # 전체 규제 강도
                          # 0.001, 0.01, 0.1, 1, 10, 100 ,1000 ...
    l1_ratio=0.5,         # L1 비율. 0이면 Ridge에 가까움, 1이면 Lasso에 가까움
    fit_intercept=True,   # 절편 학습 여부
    max_iter=10000,       # 최적화 최대 반복 횟수
    random_state=42       # 재현성 고정
)

modelElastic.fit(xTrainScale, yTrain)

pred = modelElastic.predict(xTestScale)
```

```python
# KNN Regression
# 가까운 K개의 정답 평균으로 예측하는 거리기반 모델
# 스케일링 : 권장
# 언제 사용할까
# 1. 구현하기 쉽기에 비교용으로 사용하기 좋음
# 2. 데이터의 개수가 적을 경우
# 3. 비선형 패턴이 있을 경우. 곡선 패턴도 적용 가능

from sklearn.neighbors import KNeighborsRegressor

modelKnnReg = KNeighborsRegressor(
    n_neighbors=5,        # 가까운 이웃 개수. 너무 작으면 과적합, 너무 크면 과소적합
    weights="distance",   # uniform은 동일 가중치, distance는 가까울수록 더 큰 가중치
    metric="minkowski",   # 거리 계산 방식. 기본은 minkowski
    p=2                   # p=1이면 맨해튼 거리, p=2이면 유클리드 거리
)

modelKnnReg.fit(xTrainScale, yTrain)

pred = modelKnnReg.predict(xTestScale)
```

```python
# SVR ( Support Vector Machine Regression )
# 주어진 데이터 포인트들을 가장 잘 설명하는 회귀선(초평선)을 찾음
# 설정한 오차 범위내에 최대한 많은 데이터를 포함시켜 어느정도 발생하는 오차는 봐줌
# 스케일링 : 권장
# 언제 사용할까
# 1. 비선형관계가 있을 경우
# 2. 정교한 경계가 필요한 경우

from sklearn.svm import SVR

modelSvr = SVR(
    kernel="rbf",         # rbf는 가장 자주 쓰는 비선형 커널
                          # 커널 : 2차원 평면에서 직선으로 회귀선을 긋기 힘들 때 고차원을 계산
    C=1.0,                # 규제 강도. 클수록 오차를 덜 허용하고 복잡해짐
    epsilon=0.1,          # 이 값 이하의 오차는 무시
    gamma="scale"         # rbf 영향 범위. scale을 기본으로 많이 사용
)

modelSvr.fit(xTrainScale, yTrain)

pred = modelSvr.predict(xTestScale)
```

```python
# Decision Tree Regressor
# 조건문을 이용해서 데이터를 나누고 각 구간의 평균값으로 예측함
# 스케일링 : 필요없음
# 언제 사용할까
# 1. 모델의 구조에 대해 설명해야 할 경우
# 2. 비선형 관계가 있을 경우
# 3. 변수간 중요도를 확인하고 싶을 경우

from sklearn.tree import DecisionTreeRegressor

modelTreeReg = DecisionTreeRegressor(
    max_depth=5,              # 트리 최대 깊이. 클수록 복잡해지고 과적합 위험 증가
    min_samples_split=10,     # 노드를 나누기 위한 최소 샘플 수
    min_samples_leaf=5,       # 마지막 리프 노드에 필요한 최소 샘플 수
    random_state=42           # 재현성 고정
)

modelTreeReg.fit(xTrain, yTrain)

pred = modelTreeReg.predict(xTest)
```

```python
# Random Forest Regressor
# 여러개의 Decision Tree를 만든 후, 그 평균을 최종 예측값으로 사용함
# 스케일링 : 필요없음
# 언제 사용할까
# 1. 기본적으로 성능 좋은 모델이 필요할 경우
# 2. 과적합을 줄이고 싶을 경우
# 3. 변수간 중요도를 확인하고 싶을 경우

from sklearn.ensemble import RandomForestRegressor

modelForestReg = RandomForestRegressor(
    n_estimators=300,         # Tree 개수. 많을수록 안정적이지만 느려짐
    max_depth=10,             # 각 Tree의 최대 깊이. 너무 깊으면 과적합
    min_samples_split=10,     # 노드 분할 최소 샘플 수
    min_samples_leaf=3,       # 리프 노드 최소 샘플 수
    max_features="sqrt",      # 각 Tree가 사용할 컬럼 수. sqrt, log2가 있음
    random_state=42,          # 재현성 고정
    n_jobs=-1                 # 모든 CPU 코어 사용
)

modelForestReg.fit(xTrain, yTrain)

pred = modelForestReg.predict(xTest)
```

```python
# Gradient Boosting Classifier
# 이전의 모델이 틀린 부분을 다음 모델이 계속 보완하는 트리 방식
# 스케일링 : 필요없음
# 언제 사용할까
# 1. Random Forest보다 정교하게 분류하고 싶은 경우
# 2. 데이터가 아주 크지 않을 경우
# 3. 복잡한 비선형 회귀

from sklearn.ensemble import GradientBoostingClassifier

modelGbClass = GradientBoostingClassifier(
    n_estimators=300,         # Tree 개수. 많을수록 복잡해짐
    learning_rate=0.05,       # 한 Tree가 반영되는 정도. 작게 하면 안정적이지만 Tree가 더 필요
    max_depth=3,              # 각 Tree의 깊이. 보통 3~5 사용
    min_samples_split=10,     # 노드 분할 최소 샘플 수
    min_samples_leaf=3,       # 리프 노드 최소 샘플 수
    random_state=42           # 재현성 고정
)

modelGbClass.fit(xTrain, yTrain)

pred = modelGbClass.predict(xTest)
predProba = modelGbClass.predict_proba(xTest)
```

```python
# XGBoost Classifier
# Gradient Boosting을 개선한 고성능 모델로 정형 데이터 분류에서 자주 사용함
# 스케일링 : 필요없음
# 언제 사용할까
# 1. 정형데이터 분석시 높은 성능을 원할 경우
# 2. 일부 결측치가 있을 경우 자체적으로 결측 방향 처리 가능

from xgboost import XGBClassifier  # !pip install xgboost 필요

modelXgbClass = XGBClassifier(
    n_estimators=500,         # Tree 개수. 많을수록 복잡하지만 learning_rate와 같이 조절
    learning_rate=0.05,       # 한 Tree의 반영 비율. 작게 하면 안정적이지만 Tree 더 필요
    max_depth=5,              # Tree 최대 깊이. 클수록 복잡하고 과적합 위험
    subsample=0.8,            # 행 샘플링 비율. 과적합 방지에 사용
    colsample_bytree=0.8,     # 컬럼 샘플링 비율. 과적합 방지에 사용
    reg_alpha=0.0,            # L1 규제. 클수록 불필요한 가중치를 줄이는 효과
    reg_lambda=1.0,           # L2 규제. 클수록 모델을 더 단순하게 만드는 효과
    random_state=42,          # 재현성 고정
    n_jobs=-1,                # 모든 CPU 코어 사용
    eval_metric="logloss"     # 평가 지표.  이진 분류에서는 logloss를 자주 사용
)

modelXgbClass.fit(
    xTrain, yTrain, # train_test_split로 Train -> TrainTrain, Valid로 나누기
    eval_set=[(xValid, yValid)], # 검증 데이터가 있을 경우
    early_stopping_rounds=50, # 50번 동안 개선 안 되면 멈춤
    verbose=False # 학습 로그 출력 여부
)

pred = modelXgbClass.predict(xTest)
predProba = modelXgbClass.predict_proba(xTest)
```

```python
# 스케일링
# 컬럼마다 값의 범위를 비슷하게 맞추는 전처리 기법
# 트리기반의 모델은 스케일링이 큰 영향을 못 미치지만, 그 외에는 스케일링 필요

# fit_transform()은 평균과 표준편차를 새로 계산하므로 Test에는 fit_transform()쓰면 안됨
```

```python
# StandardScaler
# 각 컬럼을 평균을 0, 표준편차가 1이 되도록 변환함
# 이상치의 영향 : 있음
# 변환값 = (원본값-평균)/표준편차
# 언제 사용할까
# 일반적인 상황에서 가장 무난함

from sklearn.preprocessing import StandardScaler

scalerStandard = StandardScaler()

xTrainScale = scalerStandard.fit_transform(xTrain)
xTestScale = scalerStandard.transform(xTest)
```

```python
# MinMaxScaler
# 값을 0과 1사이로 변환함
# 이상치의 영향 : 매우 큼
# 변환값 = (원본값-최소값)/(원본값-최대값)
# 언제 사용할까
# 값의 범위를 0~1로 맞추고 싶을 경우
# 이미지 데이터(0~255)

from sklearn.preprocessing import MinMaxScaler

scalerMinMax = MinMaxScaler(
    feature_range=(0, 1)  # 변환할 값의 범위
)

xTrainScale = scalerMinMax.fit_transform(xTrain)
xTestScale = scalerMinMax.transform(xTest)
```

```python
# RobustScaler
# 중앙값과 사분위 범위를 이용해서 스케일링
# 이상치의 영향 : 적음
# 변환값 = (원본값-중앙값)/IQR      / IQR : Q3-Q1
# 언제 사용할까
# 이상치가 많을 경우
# 극단값이 많있는 경우

from sklearn.preprocessing import RobustScaler

scalerRobust = RobustScaler()

xTrainScale = scalerRobust.fit_transform(xTrain)
xTestScale = scalerRobust.transform(xTest)
```

```python
# MaxAbsScaler
# 각 컬럼의 최대 절대값이 1이 되도록 변환함(-1~1)
# 이상치 영향 : 있음
# 변환값 = 원본값 / 최대 절대값
# 언제 사용할까
# 희소 행렬 데이터(행렬의 대부분 요소가 0으로 채워진 행렬)
# 음수와 양수가 모두 있을 경우
# 0값을 유지하고 싶을 경우

from sklearn.preprocessing import MaxAbsScaler

scalerMaxAbs = MaxAbsScaler()

xTrainScale = scalerMaxAbs.fit_transform(xTrain)
xTestScale = scalerMaxAbs.transform(xTest)
```

```python
# Normalizer
# 각 행의 크기를 1로 맞춤(다른 스케일러는 열 기준임)
# (1,3,2,1,3) ->  (0.1, 0.3, 0.2, 0.1, 0.1, 0.3)
# 언제 사용할까
# 문서 길이의 차이를 줄이고 싶을 떄

from sklearn.preprocessing import Normalizer

scalerNormalizer = Normalizer(
    norm="l2"  # l1, l2, max 사용 가능
)

xTrainScale = scalerNormalizer.fit_transform(xTrain)
xTestScale = scalerNormalizer.transform(xTest)
```

```python
# PowerTransformer
# 데이터를 정규분포에 가깝게 바꿈
# 언제 사용할까
# 데이터가 심하게 한쪽으로 치우쳐 있을 때
# 이상치 영향을 완화하고 싶을 때

from sklearn.preprocessing import PowerTransformer

scalerPower = PowerTransformer(
    method="yeo-johnson",  # yeo-johnson : 0또는 음수를 포함 가능
                           # box-cox : 양수 데이터만 가능함
    standardize=True       # 변환 후 평균 0, 표준편차 1로 맞춤
)

xTrainScale = scalerPower.fit_transform(xTrain)
xTestScale = scalerPower.transform(xTest)
```

```python
# 회귀
# 실제값과 예측 값의 차이를 보는 것이 가장 중요함
# MAE ( Mean Absolute Error )
# 오차의 절대값 평균
# MAE = |실제값-예측값| 의 평균
# 평균적으로 얼마나 틀리는지 보고 싶을 경우 사용
# 작을수록 좋음
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(yTest, pred)


# MSE ( Mean Squared Error )
# 오차를 제곱한 뒤 평균냄. 제곱하면 RMSE
# MAE = (실제값-예측값)^2 의 평균
# 큰 오차에 대해 강하게 벌점을 내고 싶을 경우 사용함
# 작을수록 좋음
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(yTest, pred)

# R2
# 결정계수로, 모델이 얼마나 정답의 변동을 잘 설명하는지 나타냄
# 1에 가까울수록 좋으며, 0근처는 평균으로 예측하는 것과 비슷, 음수는 평균 예측보다 못함
from sklearn.metrics import r2_score
r2 = r2_score(yTest, pred)
```

```python
# 분류
# 실제 클래스와 예측 클래스가 얼마나 잘 맞는지 보는 것

# Confusion Matrix
# 혼동행렬로, 정답&오답 과 예측값들에 대해 한번에 확인 가능
# 구분	의미
# TP	실제 Positive를 Positive로 예측
# TN	실제 Negative를 Negative로 예측
# FP	실제 Negative를 Positive로 잘못 예측
# FN	실제 Positive를 Negative로 잘못 예측
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(yTest, pred)

report 도 있는데

# Accuracy
# 정확도, 전체 데이터 중 맞춘 비율을 보여줌
# Accuracy = 맞춘 개수 / 전체 개수
# 불균형 데이터일 경우 매우 위험하므로, 다른 지표들도 봐야 함

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(yTest, pred)


# Precision
# 정밀도, 모델이 맞다고 예측한 것 중 진짜 맞은 비율
# 내가한 예측에 대한 신뢰도. 얼마나 예측 결과를 믿을 수 있는가

from sklearn.metrics import precision_score
precision = precision_score(yTest, pred)

# Recall
# 재현율, 실제 맞은 것들 중 맞다고 예측 성공한 비율
# 실제 정답을 얼마나 놓치지 않고 찾아냈는가

from sklearn.metrics import recall_score
recall = recall_score(yTest, pred)

# F1 Score
# Precision과 Recall의 평균
# Precision과 Recall 모두 중요할 경우 사용

from sklearn.metrics import f1_score
f1 = f1_score(yTest, pred)
```
